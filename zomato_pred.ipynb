# Zomato Analysis Project

## Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2 -Create the data frame 

```python
df = pd.read_csv("Zomato data .csv")
print(df.head())
```

```python
df
```

## Lets convert the datatype of "rate" column to float and remove the denominator

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

## Summary of Data Frame

```python
df.info()
```

## Conclusion - There is no NULL value in data frame.

## Type of Resturant

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of resturant")
```

## Conclusion: The majority of people love to dine at a hotel.

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({"votes" : grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

## THe majority of restaurants received ratings

```python
plt.hist(df['rate'],bins=5)
plt.title("Rating Distribution")
plt.show()
```

## Conclusion: The majority of restaurants received reatings ranging from 3.5 to 4.

## The majority of couples prefer restaurants with an approximate cost of 300 rupees.

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

## whether online orders receive higher ratings than offline orders.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = df)
```

## Conclusion: Online orders obtained excellent ratings as compared to Offline orders as compared.

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size')
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
